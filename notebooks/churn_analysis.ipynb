```python
1. Title (Markdown Cell)

# Customer Churn Analysis

This notebook performs data cleaning and exploratory data analysis (EDA) on telecom customer data to identify churn patterns and business insights.

## Business Objective
The goal is to identify factors affecting customer churn and provide insights to reduce revenue loss

##🔹 2. Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# for better visuals
sns.set(style="whitegrid")
##🔹 3. Load Data

data = pd.read_csv("data/raw_data.csv")
data.head()
##🔹 4. Data Understanding
data.shape

data.info()

data.describe()
##🔹 5. Check Missing Values

data.isnull().sum()
##🔹 6. Data Cleaning (IMPORTANT)

# Replace blank values with NaN
data['TotalCharges'] = data['TotalCharges'].replace('', np.nan)

# Convert to numeric
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'])

# Drop missing values
data = data.dropna(subset=['TotalCharges'])

# Verify
data['TotalCharges'].isnull().sum()
##🔹 7. Feature Engineering

def tenure_group(x):
    if x <= 12:
        return "0-1 Year"
    elif x <= 24:
        return "1-2 Years"
    elif x <= 48:
        return "2-4 Years"
    else:
        return "4+ Years"

data['TenureGroup'] = data['tenure'].apply(tenure_group)

data.head()
##🔹 8. Churn Distribution

data['Churn'].value_counts()

sns.countplot(x='Churn', data=data)
plt.title("Churn Distribution")
plt.show()
##🔹 9. Churn by Contract

pd.crosstab(data['Contract'], data['Churn'], normalize='index') * 100

sns.countplot(x='Contract', hue='Churn', data=data)
plt.title("Churn by Contract Type")
plt.xticks(rotation=20)
plt.show()
##🔹 10. Churn by Tenure Group

pd.crosstab(data['TenureGroup'], data['Churn'], normalize='index') * 100

sns.countplot(x='TenureGroup', hue='Churn', data=data)
plt.title("Churn by Tenure Group")
plt.show()
##🔹 11. Churn by Internet Service

pd.crosstab(data['InternetService'], data['Churn'], normalize='index') * 100

sns.countplot(x='InternetService', hue='Churn', data=data)
plt.title("Churn by Internet Service")
plt.show()
##🔹 12. Churn by Payment Method

pd.crosstab(data['PaymentMethod'], data['Churn'], normalize='index') * 100

sns.countplot(x='PaymentMethod', hue='Churn', data=data)
plt.xticks(rotation=30)
plt.title("Churn by Payment Method")
plt.show()
##🔹 13. Revenue Analysis

# Monthly revenue lost
data[data['Churn'] == 'Yes']['MonthlyCharges'].sum()

# Historical revenue lost
data.groupby('Churn')['TotalCharges'].sum()
##🔹 14. Key Insights (Markdown Cell)

## Key Insights

- Month-to-month contracts have the highest churn (~43%)
- New customers (0-1 year) churn the most (~47%)
- Fiber optic users show higher churn (~42%)
- Electronic check users have highest churn (~45%)
- Churn leads to significant revenue loss (~$2.86M historical)
```
